```python
from pycaret.classification import *
import pandas as pd

X = pd.read_parquet(r"../Dimensionality reduction/pca_reduced_data.parquet")
y = pd.read_parquet(r"../Dimensionality reduction/target_variable.parquet")
```

```python
# keeping some data for training, validation, testing in stratified manner
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.4, stratify=y, random_state=42
)

print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)

# Saving the data
X_train.to_parquet(r"../X_train.parquet")
X_test.to_parquet(r"../X_test.parquet")
y_train.to_parquet(r"../y_train.parquet")
y_test.to_parquet(r"../y_test.parquet")
```

```python
data = pd.concat([X_train, y_train], axis=1)
test = pd.concat([X_test, y_test], axis=1)

# Free up memory
del X, y, X_train, X_test, y_train, y_test
```

```python
import os

import mlflow.server


os.environ["MLFLOW_TRACKING_URI"] = "../../mlruns"
```

```python
# initializing setup
exp = setup(
    data=data,
    target=-1,
    train_size=0.7,
    test_data=test,
    preprocess=True,
    remove_outliers=True,
    outliers_method="iforest",
    outliers_threshold=0.05,
    fix_imbalance=True,
    fix_imbalance_method="SMOTE",
    data_split_stratify=True,
    fold_strategy="stratifiedkfold",
    fold=5,
    fold_shuffle=False,
    n_jobs=-1,
    session_id=42,
    use_gpu=True,
    log_experiment=True,
    experiment_name="model_selection_explore",
    log_plots=True,
    log_data=True,
    verbose=True,
)
```

```python
# save the experiment
exp.save_experiment(r"..\model_selection_explore")
```

```python
# exp = load_experiment(r'..\model_selection_explore',data=data,test_data=test,preprocess_data=True)
```

```python
models()
```

```python
res = exp.compare_models(
    round=4,
    sort="AUC",
    errors="raise",
    fold=2,
    include=[
        "dummy",
        "lr",
        "dt",
        "nb",
        "svm",
        "ridge",
        "rf",
        "et",
        "xgboost",
        "lightgbm",
        "catboost",
    ],
)
```

```python
# print best model and its parameters
print(res)
```

```python
exp.save_model(res, model_name="rf")
```

```python
# We have decided to use xgboost for the final model due to good performance and speed of training/inference
xg = exp.create_model("xgboost", fold=5, cross_validation=True, round=4, verbose=True)
```

```python
plot_model(
    xg,
    plot="threshold",
    save=True,
    fold=5,
)
```

```python
xg = load_model("xgboost")
```

```python
xg_tuned = tune_model(
    xg,
    fold=5,
    n_iter=5,
    optimize="F1",
    verbose=True,
    choose_better=True,
    return_train_score=True,
)
```

```python
# predict on test data and get the metrics
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

df_result = pd.DataFrame(columns=["Threshold", "Accuracy", "Precision", "Recall", "F1"])
for i in [
    0.05,
    0.1,
    0.15,
    0.2,
    0.25,
    0.3,
    0.35,
    0.4,
    0.45,
    0.5,
    0.55,
    0.6,
    0.65,
    0.7,
    0.75,
    0.8,
    0.85,
    0.9,
    0.95,
    1,
]:
    pred = predict_model(xg, data=test, probability_threshold=i)
    acc = accuracy_score(pred["Heart Attack Diagnosed"], pred["prediction_label"])
    prec = precision_score(pred["Heart Attack Diagnosed"], pred["prediction_label"])
    rec = recall_score(pred["Heart Attack Diagnosed"], pred["prediction_label"])
    f1 = f1_score(pred["Heart Attack Diagnosed"], pred["prediction_label"])
    df_result = pd.concat(
        [
            df_result,
            pd.DataFrame(
                [[i, acc, prec, rec, f1]],
                columns=["Threshold", "Accuracy", "Precision", "Recall", "F1"],
            ),
        ],
        axis=0,
    )

df_result
```

```python
# plot the df_result
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.plot(df_result["Threshold"], df_result["Accuracy"], label="Accuracy")
plt.plot(df_result["Threshold"], df_result["Precision"], label="Precision")
plt.plot(df_result["Threshold"], df_result["Recall"], label="Recall")
plt.plot(df_result["Threshold"], df_result["F1"], label="F1")
plt.legend()
plt.xlabel("Threshold")
plt.ylabel("Metrics")
plt.title("Metrics vs Threshold")
plt.show()
```
